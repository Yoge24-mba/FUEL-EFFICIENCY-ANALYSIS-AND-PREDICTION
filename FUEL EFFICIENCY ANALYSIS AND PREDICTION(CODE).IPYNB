## LOADING DATA ##

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from ISLP.models import ModelSpec as MS
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
Auto = pd.read_csv("Auto.csv")

# Replace '?' with NaN (if present)
Auto.replace('?', np.nan, inplace=True)

# Convert horsepower to numeric (if needed)
Auto['horsepower'] = pd.to_numeric(Auto['horsepower'])

# Drop missing values
Auto = Auto.dropna()

# View first rows
print(Auto.head())
```

```python
# Select predictors and response
X = Auto[['horsepower', 'weight', 'displacement']]
y = Auto['mpg']
```

```python


# Fit OLS regression
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())
```

```python
# New observation
new_car = pd.DataFrame({
    'horsepower': [110],
    'weight': [2600],
    'displacement': [160]
})
# Predict MPG
predicted_mpg = model.predict(new_car)

print("Predicted MPG:", predicted_mpg.iloc[0])
```

```python
# Predictions on training data
y_pred = model.predict(X)

# Calculate metrics
from sklearn.metrics import mean_squared_error, r2_score

print("R2 Score:", r2_score(y, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred)))
```
